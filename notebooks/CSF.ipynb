```python
# Create an image with the CSF (Contrast Sensitivity Function).

import numpy as np
import matplotlib.pyplot as plt

width = 1920
height = 1080
frequency_scaling = 0.000001
filename = "CSF.png"

# Create a grid of (X, Y) coordinates
x = np.linspace(0, width/2, width)
y = np.linspace(height-1, 0, height)
y = 1/(y/50+1)
X, Y = np.meshgrid(x, y)

# Calculate the pixel values (Z) based on the sine(x) function
# and linearly with Y
Z = Y * (1-np.sin(frequency_scaling * X**3))

plt.figure(figsize=(width / 100, height / 100), dpi=100)
plt.imshow(Z, cmap='gray', origin='upper')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
plt.savefig(filename)
plt.close()
print(f"Image saved as {filename}")
```
